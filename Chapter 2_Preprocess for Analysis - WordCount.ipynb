```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
import folium
from collections import Counter
```

```python
# Read dataset for wordcount
df = pd.read_csv('tweets_analysed_cons.csv')
df_cluster = pd.read_csv('users_clustered.csv')
df
```

```python
# Add cluster to the main dataset
chosen_cluster = 'shc_clusters'
df_cluster[chosen_cluster]

df = pd.merge(df, df_cluster[['id', chosen_cluster]], left_on='author_id', right_on='id', how='left', suffixes=['','_cluster'])
df[chosen_cluster] = df[chosen_cluster].fillna(df[chosen_cluster].max()+1)
df
```

```python
# Generate function to count words
def word_count(woc):
    return Counter(woc.split())
```

```python
# Create function to generate clean date
def clean_date(date):
    return date.rsplit(' ',3)[0]
```

```python
# Data stagging for visualisation
df_woc = df[['Sentiment', 'created_at', 'text_clean', chosen_cluster]]
df_woc['created_at_group'] = df_woc.apply(lambda x: clean_date(x.created_at), axis=1)
df_woc['text'] = df_woc[['Sentiment','text_clean','created_at_group', chosen_cluster]].groupby(['Sentiment', 'created_at_group', chosen_cluster])['text_clean'].transform(lambda x: ' '.join(x))
df_woc = df_woc[['Sentiment', 'created_at_group', 'text', chosen_cluster]].drop_duplicates()
df_woc
```

```python
# Count word on text
df_woc['woc'] = df_woc.apply(lambda x: word_count(x.text), axis=1)
df_woc
```

```python
# Create function that generate word count for a day
def woc_to_dataframe(woc, date, sentiment, cluster):
    words_ = []
    counts_ = []
    dates_ = []
    sentiments_ = []
    clusters_ = []
    for key in woc.keys():
        words_.append(key)
        counts_.append(woc[key])
        dates_.append(date)
        sentiments_.append(sentiment)
        clusters_.append(cluster)
    result = {
        'date': dates_,
        'word': words_,
        'count': counts_,
        'sentiment': sentiments_,
        'cluster': clusters_
    }
    
    return pd.DataFrame(result)
```

```python
# Transform the WordCount to dataframe
df_result = pd.DataFrame()

for index_ in range(0, len(df_woc)):
    df_new = woc_to_dataframe(df_woc.iloc[index_]['woc'],df_woc.iloc[index_]['created_at_group'], df_woc.iloc[index_]['Sentiment'], df_woc.iloc[index_][chosen_cluster])
    df_result = pd.concat([df_result, df_new])
    
df_result
```

```python
# Export wordcloud
df_result.to_csv('word_cloud.csv')
```
